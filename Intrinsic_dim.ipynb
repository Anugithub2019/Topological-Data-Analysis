```python
def plot_data(X,t):
  """ Useful function to plot data in 3D"""
  fig = plt.figure(figsize = (14,4))


  gs = GridSpec(1,1)
  ax = fig.add_subplot(gs[0], projection = '3d')
  ax.scatter(X[:, 0], X[:, 1], X[:, 2], s = 0.1, c=t, cmap= 'jet')
  ax.set_xlabel('x', fontsize = 15)
  ax.set_ylabel('z', fontsize = 15)
  ax.set_zlabel('y', fontsize = 15)
  ax.view_init(10, 105)
  gs.tight_layout(fig, rect = [0.0, 0.0, 0.21, 1])

  gs = GridSpec(1,1)
  ax = fig.add_subplot(gs[0])
  ax.scatter(-X[:, 0], X[:, 2], s = 0.1, c=t, cmap= 'jet')
  ax.set_xlabel('x', fontsize = 15)
  ax.set_ylabel('z', fontsize = 15)
  gs.tight_layout(fig, rect = [0.23, 0.05, 0.44, 0.8])


  gs = GridSpec(1,1)
  ax = fig.add_subplot(gs[0])
  ax.scatter(-X[:, 0], X[:, 1], s = 0.1, c=t, cmap= 'jet')
  ax.set_xlabel('x', fontsize = 15)
  ax.set_ylabel('y', fontsize = 15)

  gs.tight_layout(fig, rect = [0.5, 0.0, 0.71, 1])


```

```python
from sklearn.datasets import make_swiss_roll
X, t = make_swiss_roll(n_samples = 5000, noise = 0.0)
print(f'N sample ={X.shape[0]}')
plot_data(X,t)
```
